<h3>DataFrame</h3>
Ein DatenFrame ist eine Art Zweidimensionaler-Array -> Eine Tabelle mit Zeilen und Spalten. <br>
Um bestimmte Daten anzusprechen wird das Attribut <b>loc</b> verwendet. <br>
Mit dem Attribut <b>index</b> können den einzelnen Spalten Namen zugewiesen werden.

```python
# Import Pandas
import pandas as pd
```

```python
# Dictionary anlegen
Mercedes_Benz = {
    "Modell" : ["A-Klasse", "E-Klasse", "S-Klasse"],
    "Preis" : [20000, 80000, 120000]
}

# Dictionary in DatenFrame speichern und ausgeben
MB_DataFrame = pd.DataFrame(Mercedes_Benz)
print(MB_DataFrame)

# Ansprechen der ersten Zeile
print(MB_DataFrame.loc[0])
```

Mit dem Attribut <b>index</b> können den einzelnen Spalten Namen zugewiesen werden.

```python
# Bennenen der Zeilen
MB_DataFrame = pd.DataFrame(Mercedes_Benz, index = ["Auto1", "Auto2", "Auto3"])
print(MB_DataFrame)

# Ansprechen der Zeilen nach selbstgewählten Namen
print(MB_DataFrame.loc["Auto2"])
```

<h3>Analyse mithilfe von DataFrames</h3>
    - Die Methode <b>head()</b> gibt die Kopfzeilen und eine bestimmte Anzahl von Zeilen zurück, beginnend von oben. <br>
    - Die Methode <b>tail()</b> gibt die Kopfzeilen und eine bestimmte Anzahl von Zeilen zurück, beginnend von unten. <br>

```python
# Datei einlesen und Kopfzeile einbinden
temp_DataFrame = pd.read_csv("12.1 TempsSmall.csv", names=["Sensor", "Datum und Zeit", "Temperatur"])

# Head-Funktion beginnend von oben mit 10 Zeilen
print(temp_DataFrame.head(10))

# Tail-Funktion beginnend von unter mit 10 Zeilen
print(temp_DataFrame.tail(10))
```

<h2>Cleaning Data</h2>
<h3>Cleaning Empty Cells</h3>
Ein DataFrame kann alle leeren Zellen löschen, mit der Funktion <b>dropna()</b> <br>
Sollte der orginale DataFrame überschrieben werden, kann der Parameter <b>inplace</b> auf <b>true</b> gesetzt werden

```python
temp_DataFrame = pd.read_csv("12.1 TempsSmall_EmptyCells.csv")
New_temp_DataFrame = temp_DataFrame.dropna()

print(New_temp_DataFrame)

temp_DataFrame.dropna(inplace = True)

```

<h3>Replace Empty Values</h3>
Eine andere Möglichkeit, als leere Zellen zu löschen, ist diese einfach mit einem neuen Wert zu ersetzen. Mit dieser Variante müssen Sie nicht ganze Datensätze löschen, sondern können den leeren Zellen einen neuen Wert zuschreiben. Die Methode <b>fillna()</b> ersetzt alle leeren Zellen mit einem übergebenen Wert.

```python
temp_DataFrame = pd.read_csv("12.1 TempsSmall_EmptyCells.csv")

temp_DataFrame.fillna(130, inplace = True)
print(temp_DataFrame)
```

Das Beispiel oben ersetzt alle leeren Zellen im gesamten DataFrame. Um nur die leern Zellen in <u>einer</u> Spalte zu ersetzen, spezifizieren sie die Spalte. 

```python
temp_DataFrame = pd.read_csv("12.1 TempsSmall_EmptyCells.csv")

temp_DataFrame["Temperatur"].fillna(2000, inplace = True)
print(temp_DataFrame)
```

<h3>Mean, Median and Mode</h3>
Leere Zellen können auch durch einen errechneten Wert befüllt werden: <br>
- <b>mean()</b>: Der Durchschnittswert (Summe aller Werte geteilt durch die Anzahl an Werten)

```python
temp_DataFrame = pd.read_csv("12.1 TempsSmall_EmptyCells.csv")

durchschnitts_Temp = temp_DataFrame["Temperatur"].mean()
temp_DataFrame["Temperatur"].fillna(durchschnitts_Temp, inplace = True)
print(temp_DataFrame)
```

- <b>median()</b>: Der Wert in der Mitte, nachdem alle Werte aufsteigend sortiert wurden

```python
temp_DataFrame = pd.read_csv("12.1 TempsSmall_EmptyCells.csv")

durchschnitts_Temp = temp_DataFrame["Temperatur"].median()
temp_DataFrame["Temperatur"].fillna(durchschnitts_Temp, inplace = True)
print(temp_DataFrame)
```

- <b>mode()</b>: Der Wert der am meisten erscheint

```python
temp_DataFrame = pd.read_csv("12.1 TempsSmall_EmptyCells.csv")

durchschnitts_Temp = temp_DataFrame["Temperatur"].mode()
temp_DataFrame["Temperatur"].fillna(durchschnitts_Temp, inplace = True)
print(temp_DataFrame)
```

<h2>Daten im falschen Format</h2>
<h3>Konvertieren in das Korrekte Format</h3>
In unserem DataFrame haben zwei Zellen der Spalte Datum das Falsche Format (NaN). Um die Daten im falsch Format in ein DateTime-Format zu konvertieren, dies ist möglich mit der Methode <b>to_datetime()</b>


```python
temp_DataFrame = pd.read_csv("12.1 TempsSmall_EmptyCells.csv")

temp_DataFrame["Datum"] = pd.to_datetime(temp_DataFrame["Datum"])
print(temp_DataFrame)
```
