```python
# Imports
import pandas as pd
import sys
```

```python
# Einlesen der CSV-Datei
temp_DataFrame = pd.read_csv("tempsSmall.csv")
temp_DataFrame
```

```python
# Variablen initialisieren
maxTemp = sys.float_info.min
minTemp = sys.float_info.max
avgTemp = 0

# Maximale Temperatur ermitteln
for temperatur in temp_DataFrame.values:
    if temperatur[2] > maxTemp:
        maxTemp = temperatur[2]

# Minimale Temperatur ermitteln
for temperatur in temp_DataFrame.values:
    if temperatur[2] < minTemp:
        minTemp = temperatur[2]

# Durschnittliche Temperatur ermitteln
count = 0
tempSum = 0
for temperatur in temp_DataFrame.values:
    tempSum += temperatur[2]
    count += 1 

avgTemp = tempSum / count

print(f"The highest temperature is {maxTemp}")
print(f"The lowest temperature is {minTemp}")
print(f"The average temperature is {avgTemp}")
```

```python
# CSV-Datei Spalten bearbeiten
temp_DataFrame.columns = ["Sensor", "Datum", "Temperatur"]
```

```python
# Maximale Temperatur ermitteln
print(temp_DataFrame.max()[2])

# Minimale Temperatur ermitteln
print(temp_DataFrame.min()[2])
```

```python
# Maximale Temperatur ermitteln
print(temp_DataFrame["Temperatur"].max())

# Minimale Temperatur ermitteln
print(temp_DataFrame["Temperatur"].min())

# Durchschnittliche Temperatur ermitteln
print(temp_DataFrame["Temperatur"].mean())
```
